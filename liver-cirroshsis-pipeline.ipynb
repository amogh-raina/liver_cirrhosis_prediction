### 3. **Random sample imputation**

```python
df_random = df.copy()
def impute_random(df,columns):
    for i in columns:
        random_sample=df[i].dropna().sample(df[i].isnull().sum(),random_state=0) 
    ## Replacing the NaN values with some random sampled value
    ##pandas need to have same index in order to merge the dataset
        random_sample.index=df[df[i].isnull()].index
        df.loc[df[i].isnull(),i]=random_sample

impute_random(df_random,df_random.columns)
```

# **Transformations using pipelines**

```python
from sklearn.preprocessing import FunctionTransformer
from sklearn.compose import ColumnTransformer, make_column_transformer 
from sklearn.preprocessing import RobustScaler, StandardScaler, MinMaxScaler
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from sklearn.pipeline import Pipeline,make_pipeline
ohe = OneHotEncoder(sparse=False,drop='first')
le = LabelEncoder()
scaler_1 = RobustScaler()
scaler_2 = StandardScaler()
scaler_3 = MinMaxScaler()
#################################################################
from sklearn.naive_bayes import GaussianNB, BernoulliNB, CategoricalNB
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
import lightgbm as lgb
xg = XGBClassifier()
lbg_clf = lgb.LGBMClassifier()
rf_clf = RandomForestClassifier()
gnb_clf = GaussianNB()
bnb_clf = BernoulliNB()
cnb_clf = CategoricalNB()
```

```python
X_train.head()
```

```python
X_train.info()
```

```python
X_train = X_train.astype({'Age':'int','Bilirubin':'float', 'Cholesterol':'float', 'Albumin':'float', 
        'Copper':'float', 'Alk_Phos':'float', 'SGOT':'float',
       'Tryglicerides':'float', 'Platelets':'float', 'Prothrombin':'float'
})

X_valid = X_valid.astype({'Age':'int','Bilirubin':'float', 'Cholesterol':'float', 'Albumin':'float', 
        'Copper':'float', 'Alk_Phos':'float', 'SGOT':'float',
       'Tryglicerides':'float', 'Platelets':'float', 'Prothrombin':'float'
})
```

```python
numeric_transformer = Pipeline(steps=[
   ('imputer', KNNImputer(n_neighbors=10, weights="distance")),
   ('scaler', scaler_2)])

categorical_transformer = Pipeline(steps=[
   ('onehot', ohe)])

preprocessor = ColumnTransformer(
   remainder = 'passthrough',
   transformers=[
       ('numeric', numeric_transformer,[1,7, 8, 9,10,11,12,13,14,15]),
       ('categorical', categorical_transformer,[0,2,3,4,5,6])
])

transform_1 = Pipeline(
   steps=[
       ("processing", preprocessor),
       ("GaussianNB",gnb_clf),
   ]
)

transform_2 = Pipeline(
   steps=[
       ("processing", preprocessor),
       ("BernoulliNB",bnb_clf),
   ]
)

## CategoricalNB only works with positive values, so in order to use this you have to use min_max_scaler
## as min_max_scaler outputs value between 0 and 1
transform_3 = Pipeline(
   steps=[
       ("processing", preprocessor),
       ("CategoricalNB",cnb_clf),
   ]
)

transform_4 = Pipeline(
   steps=[
       ("processing", preprocessor),
       ("XgBoost",xg),
   ]
)

transform_5 = Pipeline(
   steps=[
       ("processing", preprocessor),
       ("RandomForestClassifier",rf_clf),
   ]
)

transform_6 = Pipeline(
   steps=[
       ("processing", preprocessor),
       ("LGBMClassifier",lbg_clf ),
   ]
)


```
